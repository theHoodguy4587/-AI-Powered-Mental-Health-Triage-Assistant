## Distribution of labels in training set

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
sns.countplot(x='label', data=train)    
plt.xlabel('label')
plt.ylabel('count')
plt.title('Distribution of labels in training set')
plt.show()
```
